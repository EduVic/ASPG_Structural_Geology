# APSG tutorial - structural geology module for Python - Part 6

```python
from apsg import *
import numpy as np
apsg_conf['figsize'] = (9, 7)
```

Some fun with cones...

```python
l1 = lin(110, 40)
l2 = lin(250, 40)
m = l1.slerp(l2, 0.5)
p = lin(l1.cross(l2))
```

```python
s = StereoNet()
```

```python
for t in np.linspace(0, 1, 8):
    D = defgrad.from_vectors_axis(l1, l2, m.slerp(p, t))
    a, ang = D.axisangle()
    s.cone(cone(a, l1, ang))
```

```python
for a in np.linspace(10,25,4):
    s.cone(cone(lin(315, 30), a), cone(lin(45, 30), a))
```

```python
s.show()
```
